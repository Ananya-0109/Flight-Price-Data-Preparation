```python
import numpy as np

def convert_duration(x):
    try:
        if isinstance(x, (int, float)):
            return x * 60
        elif isinstance(x, str):
            x = x.strip()
            if not x:  # Handle empty strings
                return np.nan
            hours = 0
            mins = 0
            if 'h' in x:
                hours_part = x.split('h')[0].strip()
                if hours_part:
                    hours = float(hours_part)
                x = x.split('h')[1]
            if 'm' in x:
                mins_part = x.split('m')[0].strip()
                # After splitting by 'h', if nothing before 'm', mins_part is ''. Handle this:
                if ' ' in mins_part:
                    mins_part = mins_part.split()[-1]
                if mins_part:
                    mins = float(mins_part)
            return hours * 60 + mins
        else:
            return np.nan
    except Exception as e:
        return np.nan
df['duration_mins'] = df['duration'].apply(convert_duration)
print(df[['duration', 'duration_mins']].head(10))
print(f"Number of NaN values: {df['duration_mins'].isna().sum()}")
df.head()
```

```python
import re

def clean_stops(x):
    if not isinstance(x, str):
        return x
    # Remove newline, tabs, and dots
    x = x.strip().replace('\n', '').replace('\t', '').replace('.', '')
    # Extract 'non-stop' or '[number]-stop'
    match = re.search(r'(\d+-stop|non-stop)', x)
    if match:
        return match.group(1)
    else:
        return x  # return as-is if no match

# Example usage on your DataFrame column:
df['stops_clean'] = df['stops'].apply(clean_stops)
```

```python
import re

def clean_stops(x):
    if not isinstance(x, str):
        return x
    # Remove any actual whitespace, literal "\n", "\t", and all escaped variants
    x = re.sub(r'(\\+t|\\+n|\t|\n|\.{2,})', '', x)
    x = x.strip()
    # Extract 'non-stop' or '1-stop', '2-stop', etc.
    match = re.search(r'(\d+-stop|non-stop)', x)
    if match:
        return match.group(1)
    else:
        return x

df['stops_clean'] = df['stops'].apply(clean_stops)
df.head()
```
